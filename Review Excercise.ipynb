IV.	List Comprehension

```python

Created_List = [1225, 4986, 6789, 7890, 2345, 6783, 987, 1234, 8765, 3456]

squared_list = [item ** 2 for item in Created_List]

print("List Comprehension (squared values):")
print(squared_list)
```

V.	Enumerate

```python

Created_List = [1225, 4986, 6789, 7890, 2345, 6783, 987, 1234, 8765, 3456]

print("Iterating using enumerate:")
for index, value in enumerate(Created_List):
    print(f"Index: {index}, Value: {value}")
```

VI.	Iter function and next function

```python

Created_List = [1225, 4986, 6789, 7890, 2345, 6783, 987, 1234, 8765, 3456]


iterator = iter(Created_List)

while True:
    try:
        number = next(iterator)
        print(number)
    except StopIteration:
        
        break
```

VII.	Map function

```python

Created_List = [1225, 4986, 6789, 7890, 2345, 6783, 987, 1234, 8765, 3456]


def square(x):
    return x ** 2

squared_list = list(map(square, Created_List))

print("List after applying map function (squared values):")
print(squared_list)
```

3)	For a List A, B, C, D, E write a python program to compute all the combinations and permutations 

```python

A = [1, 2]
B = [3, 4]
C = [5, 6]
D = [7, 8]
E = [9, 10]

all_lists = [A, B, C, D, E]

def combinations(lists, r):
    if r == 0:
        return [[]]
    result = []
    for i in range(len(lists)):
        for combo in combinations(lists[i+1:], r-1):
            result.append([lists[i]] + combo)
    return result

def permutations(lists, r):
    if r == 0:
        return [[]]
    result = []
    for i in range(len(lists)):
        for perm in permutations(lists[:i] + lists[i+1:], r-1):
            result.append([lists[i]] + perm)
    return result

print("Combinations:")
for r in range(1, len(all_lists) + 1):
    combos = combinations(all_lists, r)
    print(f"\nCombinations of {r} lists:")
    for combo in combos:
        print(combo)

print("\n" + "="*40 + "\n")  

print("Permutations:")
for r in range(1, len(all_lists) + 1):
    perms = permutations(all_lists, r)
    print(f"\nPermutations of {r} lists:")
    for perm in perms:
        print(perm)
```

4)	Using the same list use itertools compute permutations and combinations 

```python
import itertools

A = [1, 2]
B = [3, 4]
C = [5, 6]
D = [7, 8]
E = [9, 10]

all_lists = [A, B, C, D, E]

print("Original lists:")
for i, lst in enumerate(all_lists):
    print(f"List {chr(65+i)}: {lst}")

print("\nCombinations:")
for r in range(1, len(all_lists) + 1):
    combinations = list(itertools.combinations(all_lists, r))
    print(f"\nCombinations of {r} lists:")
    for combo in combinations:
        print(combo)

print("\n" + "="*40 + "\n")  

print("Permutations:")
for r in range(1, len(all_lists) + 1):
    permutations = list(itertools.permutations(all_lists, r))
    print(f"\nPermutations of {r} lists:")
    for perm in permutations:
        print(perm)
```
